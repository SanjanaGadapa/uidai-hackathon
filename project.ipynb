```python
final_df = (
    enroll
    .groupby(['new_dates', 'new_district', 'pincode'], as_index=False)
    .agg({
        'age_0_5': 'sum',
        'age_5_17': 'sum',
        'age_18_greater': 'sum',
        'total_enrollment': 'sum'
    })
)
final_df
```

```python
final_df.duplicated(
    subset=['new_dates','new_district','pincode'],
    keep=False
).sum()
```

```python
final_df.groupby('pincode')['new_district'].nunique().sort_values(ascending=False).head()
```

```python
final_df.isnull().sum()
```

```python
# Which age group has the highest total enrollment?

age_totals = final_df[['age_0_5','age_5_17','age_18_greater']].sum()
print(age_totals)

age_totals.plot(kind='bar')
plt.xlabel('Age Groups')
plt.ylabel('Total Enrollment')
plt.title('Total Enrollment by Age Group')
```

```python
# Top 5 districts with highest enrollment

top_districts = final_df.groupby('new_district')['total_enrollment'].sum().sort_values(ascending=False).head(5)
print(top_districts)

top_districts.plot(kind='barh')
plt.ylabel('Districts')
plt.xlabel('Total Enrollment')
plt.title('Top 5 Districts by Total Enrollment')
```

```python
# In which district is enrollment of 0–5 highest?

age_district = final_df.groupby('new_district')[['age_0_5', 'age_5_17', 'age_18_greater']].sum()

district_0_5 = age_district['age_0_5'].sort_values(ascending=False).head(3)
print(district_0_5)

district_0_5.plot(kind='barh')
plt.ylabel('Districts')
plt.xlabel('Enrollment (Age 0-5)')
plt.title('Top 3 Districts by Enrollment (Age 0-5)')
```

```python
# In which district is enrollment of 5-17 highest?

district_5_17 = age_district['age_5_17'].sort_values(ascending=False).head(3)
print(district_5_17)

district_5_17.plot(kind='barh')
plt.ylabel('Districts') 
plt.xlabel('Enrollment (Age 5-17)')
plt.title('Top 3 Districts by Enrollment (Age 5-17)')
```

```python
# In which district is enrollment of 18 and 18+ highest?

district_18_greater = age_district['age_18_greater'].sort_values(ascending=False).head(3)
print(district_18_greater)

district_18_greater.plot(kind='barh')
plt.ylabel('Districts')
plt.xlabel('Enrollment (Age 18+)')
plt.title('Top 3 Districts by Enrollment (Age 18+)')
```

```python
# Which districts contribute the highest share of total enrollments?

district_share = (final_df.groupby('new_district')['total_enrollment'].sum()/final_df['total_enrollment'].sum()) * 100
district_share.sort_values(ascending=False).head(3)

# While urban districts record higher enrollment volumes, the overall distribution of Aadhaar enrollments across Maharashtra 
# remains relatively balanced, with no single district contributing a disproportionately large share.
```

```python
age_district_pr = age_district.div(age_district.sum(axis=1), axis=0)*100
age_district_pr
```

```python
# Proportion of age groups in top 10 districts by total enrollment

top_districts = age_district.sum(axis=1).sort_values(ascending=False).head(10).index

top_age_pct = age_district_pr.loc[top_districts]

top_age_pct.plot(kind='bar', stacked=True, figsize=(12,6), colormap='viridis')

plt.ylabel('Percentage of Enrollment')
plt.xlabel('District')
plt.title('Age Group Composition of Enrollment in Top 10 Districts')
plt.legend(title='Age Group')
plt.xticks(rotation=45)
plt.show()
```

```python
# Age-wise trends over time

monthly_age = (
    final_df
    .set_index('new_dates')
    .resample('ME')[['age_0_5', 'age_5_17', 'age_18_greater']]
    .sum()
)

print(monthly_age)

monthly_age.plot(figsize=(10,5))

plt.xlabel('Month')
plt.ylabel('Total Enrollment')
plt.title('Monthly Aadhaar Enrollment Trend by Age Group')
plt.legend(title='Age Group')
plt.grid(True)
plt.show()
```

```python
# How much total enrollment happened for each pincode?

final_df.groupby(['pincode','new_district'])['total_enrollment'].sum().sort_values(ascending=False)
```

```python
pincode_spread = (
    final_df
    .groupby('new_district')['pincode']
    .nunique()
    .sort_values(ascending=False)
)

pincode_spread.head(10)
```

```python
district_total = final_df.groupby('new_district')['total_enrollment'].sum()
district_pincodes = final_df.groupby('new_district')['pincode'].nunique()

enrollment_per_pincode = (district_total / district_pincodes).sort_values(ascending=False)

enrollment_per_pincode.head(10)
```

```python
district_access_concentration = pd.DataFrame({
    'total_enrollment': district_total,
    'unique_pincodes': district_pincodes,
    'enrollment_per_pincode': enrollment_per_pincode
}).sort_values(by='total_enrollment', ascending=False)

district_access_concentration.head(10)
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=district_access_concentration,
    x='unique_pincodes',
    y='total_enrollment'
)

plt.xlabel('Number of Unique Pincodes')
plt.ylabel('Total Enrollment')
plt.title('District-wise Enrollment Spread vs Volume')

plt.show()
```

```python
from scipy.stats import zscore

district_total = final_df.groupby('new_district')[['age_0_5','age_5_17','age_18_greater']].sum()
district_total['total'] = district_total.sum(axis=1)
district_total['z_score'] = zscore(district_total['total'])

# Highlight outliers (z > 2 or z < -2)
outliers = district_total[district_total['z_score'].abs() > 2]
print("Outlier Districts based on total enrollment:\n", outliers[['total','z_score']])
```
